```python
# Function to calculate grade based on average marks
def calculate_grade(avg):
    if avg >= 90:               # If average is 90 or more
        return "A"              # Grade is A
    elif avg >= 75:             # If average is 75 or more (but less than 90)
        return "B"              # Grade is B
    else:                       # If average is less than 75
        return "C"              # Grade is C

# Input marks for 3 subjects from the user using list comprehension
# 'i' goes from 0 to 2 (range(3)), so it asks 3 times
marks = [int(input(f"Enter mark {i+1}: ")) for i in range(3)]

# Calculate total marks by summing the list
total = sum(marks)

# Calculate average marks
avg = total / len(marks)  # len(marks) = number of subjects (3 in this case)

# Print total marks
print("Total:", total)

# Print average marks
print("Average:", avg)

# Call the function calculate_grade with average marks
# and print the result
print("Grade:", calculate_grade(avg))
```
